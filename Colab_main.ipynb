```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_activity_box(ax, x, y, activity, duration, es, ef, ls, lf):
    """Aktivite kutusu çizimi"""
    width = 2
    height = 2

    # Ana kutu
    rect = patches.Rectangle((x, y), width, height, facecolor='white', edgecolor='black', linewidth=1)
    ax.add_patch(rect)

    # İç çizgiler
    ax.plot([x, x+width], [y+height/2, y+height/2], 'black', linewidth=1)  # Yatay çizgi
    ax.plot([x+width/2, x+width/2], [y, y+height], 'black', linewidth=1)   # Dikey çizgi

    # Değerler
    ax.text(x+width/4, y+height*3/4, str(es), ha='center', va='center', fontsize=12)
    ax.text(x+width*3/4, y+height*3/4, str(ef), ha='center', va='center', fontsize=12)
    ax.text(x+width/4, y+height/4, str(ls), ha='center', va='center', fontsize=12)
    ax.text(x+width*3/4, y+height/4, str(lf), ha='center', va='center', fontsize=12)

    # Aktivite adı ve süre
    if activity:
        ax.text(x+width/2, y+height+0.3, f"{activity}({duration})", ha='center', va='bottom', fontsize=12)

def draw_cpm_diagram():
    """CPM Diyagramı Çizimi"""
    fig, ax = plt.subplots(figsize=(20, 8))

    # Aktiviteler ve koordinatları
    activities = {
        'PPA': {'x': 4, 'y': 3, 'es': 0, 'ef': 2, 'ls': 0, 'lf': 2, 'dur': 2},
        'ST': {'x': 8, 'y': 5, 'es': 2, 'ef': 6, 'ls': 2, 'lf': 6, 'dur': 4},
        'VTT': {'x': 8, 'y': 1, 'es': 2, 'ef': 4, 'ls': 2, 'lf': 4, 'dur': 2},
        'FG': {'x': 12, 'y': 5, 'es': 6, 'ef': 9, 'ls': 6, 'lf': 9, 'dur': 3},
        'BG': {'x': 12, 'y': 1, 'es': 4, 'ef': 9, 'ls': 4, 'lf': 9, 'dur': 5},
        'TS': {'x': 16, 'y': 5, 'es': 9, 'ef': 12, 'ls': 9, 'lf': 12, 'dur': 3},
        'GT': {'x': 16, 'y': 1, 'es': 9, 'ef': 11, 'ls': 9, 'lf': 11, 'dur': 2},
        'KKT': {'x': 20, 'y': 3, 'es': 12, 'ef': 14, 'ls': 12, 'lf': 14, 'dur': 2},
        'PY': {'x': 24, 'y': 3, 'es': 14, 'ef': 15, 'ls': 14, 'lf': 15, 'dur': 1}
    }

    # Başlangıç ve bitiş düğümleri
    circle_start = plt.Circle((2, 4), 0.8, fill=False)
    circle_end = plt.Circle((28, 4), 0.8, fill=False)
    ax.add_artist(circle_start)
    ax.add_artist(circle_end)
    ax.text(2, 2.8, 'BAŞLANGIÇ', ha='center', va='top', fontsize=12)
    ax.text(28, 2.8, 'BİTİŞ', ha='center', va='top', fontsize=12)

    # Aktivite kutuları
    for act, props in activities.items():
        draw_activity_box(ax, props['x'], props['y'], act, props['dur'],
                        props['es'], props['ef'], props['ls'], props['lf'])

    # Oklar
    arrows = [
        ('PPA', 'ST'), ('PPA', 'VTT'),
        ('ST', 'FG'), ('VTT', 'BG'),
        ('FG', 'TS'), ('BG', 'GT'),
        ('TS', 'KKT'), ('GT', 'KKT'),
        ('KKT', 'PY')
    ]

    for start, end in arrows:
        start_x = activities[start]['x'] + 2
        start_y = activities[start]['y'] + 1
        end_x = activities[end]['x']
        end_y = activities[end]['y'] + 1

        # Ok çizgisi
        ax.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y),
                   arrowprops=dict(arrowstyle='->', linewidth=1))

    # Başlangıç ve bitiş okları
    ax.annotate('', xy=(activities['PPA']['x'], activities['PPA']['y']+1),
               xytext=(2.8, 4), arrowprops=dict(arrowstyle='->', linewidth=1))
    ax.annotate('', xy=(28, 4), xytext=(activities['PY']['x']+2,
               activities['PY']['y']+1), arrowprops=dict(arrowstyle='->', linewidth=1))

    # Grafik ayarları
    ax.set_xlim(0, 30)
    ax.set_ylim(0, 8)
    ax.axis('off')
    plt.title('CPM Kritik Yol Analizi', pad=20, fontsize=14)
    plt.show()

# Diyagramı çiz
draw_cpm_diagram()
```
